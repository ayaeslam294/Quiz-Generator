```python
from PyPDF2 import PdfReader
import docx
import os
from google.colab import files
import random
from IPython.display import HTML, display

# ===== ÙˆØ§Ø¬Ù‡Ø© ØªØ±Ø­ÙŠØ¨ + Ø±ÙØ¹ Ø§Ù„Ù…Ù„Ù =====
display(HTML('''
<div style="
    background-color:white;
    color:black;
    text-align:center;
    font-size:28px;
    font-weight:bold;
    padding:20px;
    border-radius:15px;
    border:2px solid #000;
    width:80%;
    margin:auto;
    margin-top:30px;
">
    Welcome to Aya's AI Quiz Generator ğŸ¤–
    <p style="font-size:18px; font-weight:normal;">
        Please upload your file below (TXT / PDF / DOCX)
    </p>
</div>
'''))

# ===== ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ù…Ù„Ù =====
uploaded = files.upload()
file_path = list(uploaded.keys())[0]
print("Uploaded:", file_path)

# ===== Ø¯ÙˆØ§Ù„ Ù‚Ø±Ø§Ø¡Ø© Ø§Ù„Ù…Ù„ÙØ§Øª =====
def read_txt(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        return f.read()

def read_pdf(file_path):
    reader = PdfReader(file_path)
    text = ""
    for page in reader.pages:
        page_text = page.extract_text()
        if page_text:
            text += page_text + "\n"
    return text

def read_docx(file_path):
    doc = docx.Document(file_path)
    text = ""
    for para in doc.paragraphs:
        text += para.text + "\n"
    return text

# ===== ØªÙˆÙ„ÙŠØ¯ Ø§Ù„Ø£Ø³Ø¦Ù„Ø© =====
def generate_questions_offline(text, num_questions):
    sentences = [s.strip() for s in text.split('.') if len(s.strip()) > 10]
    questions = []

    for i in range(min(num_questions, len(sentences))):
        sentence = sentences[i]
        question = f"Q{i+1}: {sentence}?"
        correct_answer = sentence.split()[0]

        words = list(set(" ".join(sentences).split()))
        wrong_answers = random.sample(words, 3) if len(words) > 3 else ["X", "Y", "Z"]

        options = [correct_answer] + wrong_answers
        random.shuffle(options)

        questions.append({
            "question": question,
            "options": options,
            "answer": correct_answer
        })
    return questions

# ===== ØªÙ†ÙÙŠØ° Ø§Ù„Ø¨Ø±Ù†Ø§Ù…Ø¬ =====
if file_path.endswith(".txt"):
    content = read_txt(file_path)
elif file_path.endswith(".pdf"):
    content = read_pdf(file_path)
elif file_path.endswith(".docx"):
    content = read_docx(file_path)
else:
    print("Unsupported file format.")
    content = None

if content:
    num_q = int(input("How many questions do you want? (10 / 20 / 30): "))
    questions = generate_questions_offline(content, num_q)

    print("\n=== Quiz ===\n")
    for q in questions:
        print(q["question"])
        for idx, opt in enumerate(q["options"]):
            print(f"{chr(65+idx)}. {opt}")
        print()

    show_ans = input("Press 0 if you want to see the answers: ")
    if show_ans.strip() == "0":
        print("\n=== Correct Answers ===\n")
        for q in questions:
            print(f"Q{questions.index(q)+1}: {q['answer']}")
```
